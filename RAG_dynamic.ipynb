```python
from langchain_openai import ChatOpenAI
from langchain_ollama import ChatOllama

llm = ChatOllama(model="llama3")
```

```python

llm_local = ChatOpenAI(
    base_url="http://localhost:12434/v1",
    api_key="not-needed",
    model="llama3.1",
    temperature=0.9
)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader

loader = PyMuPDFLoader("./VectorStore/enterprise_sales_report.pdf")

docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)

chunks = splitter.split_documents(docs)
```

```python
from langchain_ollama import OllamaEmbeddings

embedding_model = OllamaEmbeddings(
    model="nomic-embed-text"
)
```

```python
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    
    documents=chunks,
    embedding=embedding_model

)
```

```python
print("\n System Ready! Ask your questions (type 'exit' to quit)\n")

while True:
    query = input("Enter your question: ")

    if query.lower() == "exit":
        print(" Exiting...")
        break

    context = vectorstore.similarity_search(query, k=4)

    response = llm.invoke( f"Answer the question based only on the context below:\n\n{context}\n\nQuestion: {query}")

    print("\n Answer:", response.content, "\n")
```
